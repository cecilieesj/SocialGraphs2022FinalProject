## Reading the text pages

```python
# Importing the character list
t = os.listdir("completelist/")
disney = [w[0:-4] for w in t]
```

```python
# Checking the length of the list
len(disney)
```

```python
# Create empty graph
G = nx.DiGraph()
```

```python
# Creating the network based on the list
i = 0
for character in t:
    filename = "completelist/" + character
    string = open(filename, "r", encoding="utf-8").read()
    pattern = r'\[\[(.+?)\]\]'
    
    #find links in text
    match = re.findall(pattern, string)
    new_match = []
    for m in match:
        new_name = m.split('|')[0]
        new_name = new_name.split('#')[0]
        new_match.append(new_name)
    
    #find number of words in text
    words = string.split()
    
    character = character[0:-4]
    i = i+1
    print(character," ", i)

    G.add_node(character)
    for char in t:
        char = char[0:-4] 
        for link in new_match:
            if (char == link):
                if (char not in list(G.nodes())):
                    G.add_node(char)    
                G.add_edge(character, char)
```

```python
print("The number of nodes in the network G is: " + str(len(G.nodes())))
print("The number of edges in the network G is: " + str(len(G.edges())))
```

```python
# Remove nodes without links
for char in list(G.nodes()):
    if (G.degree[char] == 0):
        G.remove_node(char)
```

```python
print("The number of nodes in the network G is: " + str(len(G.nodes())))
print("The number of edges in the network G is: " + str(len(G.edges())))
```

```python
# Get largest connected component
largest_cc = max(nx.weakly_connected_components(G))
G = G.subgraph(largest_cc)
```

```python
print("The number of nodes in the network G is: " + str(len(G.nodes())))
print("The number of edges in the network G is: " + str(len(G.edges())))
```

```python
# Remove node with no name
for node in list(G.nodes()):
    if (node == ''):
        G.remove_node(node)
```

```python
# Create edgelist of the network
nx.write_edgelist(G, "G.edgelist", delimiter='\t')
```

```python
# Checking that the reding of the edgelist works
G_disney = nx.read_edgelist("G.edgelist", delimiter='\t', create_using=nx.DiGraph) 
```

## Creating attributes to the network

### Gender, good_vs_bad and universe attributes

```python
# Reading the gender lists
filename = "Females.txt"
females = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")

filename = "Males.txt"
males = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")
```

```python
# Reading the good_vs_bad lists
filename = "Heroines.txt"
heroine = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")

filename = "Heroes.txt"
hero = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")

filename = "Villains.txt"
villain = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")
```

```python
# Reading the universe lists
filename = "Disney_characters.txt"
disney = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")

filename = "Pixar_characters.txt"
pixar = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")

filename = "Marvel_Comics_characters.txt"
marvel = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")

filename = "Star_Wars_characters.txt"
star_wars = open(filename, "r", encoding="utf-8").read().replace("_", " ").split("|||")
```

```python
# Making lists with the characters attributes
gender = []
good_vs_bad = []
universe = []

for character in list(G.nodes()):
    # gender attribute
    if character in females:
        gender.append("Female")
    elif character in males:
        gender.append("Male")
    else:
        gender.append("Other")
        
    # God vs bad attribute  
    if character in heroine:
        good_vs_bad.append("Heroine")
    elif character in hero:
        good_vs_bad.append("Hero")
    elif character in villain:
        good_vs_bad.append("Villain")
    else:
        good_vs_bad.append("Neutral")
        
        
    # universe attribute
    if character in disney:
        universe.append("Disney")
    elif character in pixar:
        universe.append("Pixar")
    elif character in marvel:
        universe.append("Marvel")
    elif character in star_wars:
        universe.append("Star Wars")
    else:
        universe.append("Other")
```

### Movie attribute

```python
# Detecting the films from each character and crating a list
films = []
for character in list(G.nodes()):
    print(character)
    filename = "completelist/" + character + ".txt"
    string = open(filename, "r", encoding="utf-8").read().split("\n")
    
    film = "None"
    for line in string:
        if line[:5] == "|film":
            l = line.split("<br>")[0]
            if "[[" in str(l):
                l = l.split("]]")[0]
                l = l.split("[[")[1]
                l = l.split("|")[0]
                l = l.split(" (")[0]
            
            if "''" in str(l):
                l = l.split("''")[1]
                
            film = l.strip()
        
    films.append(film)
```

```python
# Removing films that are empty and set them to "None"
for i in range(0,len(films)):
    if films[i] == "|films =" or films[i] == "|films=" or films[i] == "|films = " or films[i] == "|films= " or films[i] == "|films":
        films[i] = "None" 
        
    if films[i] == "|films=|shorts=|games=|rides=|animator=Dan Santat|designer=|actor=|actress=|model=|inspiration=|awards=|fullname=|alias=|personality=|appearance=Long brown hair, baby blue eyes, necklace, hot pink jacket, blue graphic tee, white pants, pink flats|alignment=|affiliations=|goal=|home=|family=|pets=|friends=|minions=|enemies=|powers=|paraphernalia=|weapons=|possessions=|fate=|quote=}}":
        films[i] = "None"
        
    if films[i] == "|films=|shorts=|games=|rides=|animator=|designer=|voice=Ron Pardo|actor=|actress=|model=|inspiration=|awards=|fullname=|alias=|appearance=|affiliations=|home=|family=|pets=|weapons=|possessions=|fate=}}":
        films[i] = "None"

# Manually removing part of strings that are incorrect
    if films[i] == '|films = Handy Manny: The Movie' or films[i] == '|films = Liv and Maddie':
        films[i] = films[i].split("= ")[1]
        
    if films[i] == '|films =Captain America: The First Avenger':
        films[i] = films[i].split("=")[1]
        
    if 'w:c:marvelcinematicuniverse: ' in films[i]:
        films[i] = films[i].split(": ")[1]
        
    if 'w:c:muppet:' in films[i] or 'w:c:muppets:' in films[i] or  'w:c:Muppet:' in films[i] or 'W:c:muppet:' in films[i] or 'w:c:marvelcinematicuniverse:' in films[i] or 'w:c:marvelcineaticuniverse:' in films[i]:
        films[i] = films[i].split(":")[3]
        
    if 'Wikipedia:' in films[i]:
        films[i] = films[i].split(":")[1]
        
    if films[i] == '{{WikipediaLink|The Muppet Movie}}':
        films[i] = 'The Muppet Movie'     
```
